# Introduction
Last week, we saw how to model repeated measurements probabilistically using the *multivariate* normal distribution. Key to this, was the specification of the *variance-covariance matrix*, which captures the pattern of correlation across the repeated measurements. We also saw that one of the problems with the repeated measures ANOVA was that it assumed a very restrictive covariance structure in the form of *compound symmetry*. This had direct consequences for inference because the covariance structure directly informs the standard errors and thus the denominator of the tests statistics, $p$-values and confidence intervals. As such, our concern around modelling repeated measures corresponds directly to the use of a suitable covariance structure to represent the data-generating process. In an ideal world, what we want is a method that makes *no* assumptions about the covariance structure and just allows it to be estimated from the data. Unfortunately, this desire has consequences for inference. In short, the condition of compound symmetry is imposed for the repeated measures ANOVA *precisely* because that is the one situation where inference still works. As soon as we move away from this to an *arbitrary covariance structure*, the inferential machinery *falls apart*. In this lesson, we will begin by focusing on this issue, because it has direct consequences for all the methods we will cover going forward. We will then discuss GLS as one potential approach that allows for a more general covariance structure, again focusing on how inference can be conducted under these conditions. This will pave the way for understanding the full mixed-effects framework, which will be our focus *next week*.
